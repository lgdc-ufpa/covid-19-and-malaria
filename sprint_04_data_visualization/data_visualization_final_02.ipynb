```python
df_mean_daily_deaths_2020 = load_mean_daily_deaths_2020()

load_mean_daily_deaths_2020.index = df_mean_daily_deaths_2020.index.astype('str')

df_mean_daily_deaths_2020
```

```python
df_cfc_2020 = load_cfc_2020().set_index('codmun')

df_cfc_2020.index = df_cfc_2020.index.astype('str')

df_cfc_2020
```

```python
df_covid_data_to_correlact = df_mean_daily_deaths_2020.join(df_cfc_2020)
df_covid_data_to_correlact
```

```python
corr = df_covid_data_to_correlact.corr(method='spearman')

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

plt.figure(figsize = (10,10))

sns.heatmap(data=corr, cmap=cmap, mask=mask, center=0, vmax=1, vmin=-1, annot=True)
plt.title('Spearman Correlation', size=18,fontweight="bold")

path_output = os.path.join(os.getcwd(), 'output')

plt.savefig(os.path.join(path_output, 'final_correlation_covid_only.png'), dpi=300)
```
