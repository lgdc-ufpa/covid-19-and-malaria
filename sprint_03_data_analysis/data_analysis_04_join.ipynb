# <center>Join dataframes df_covid and df_indicators_2</center>

## 1. Imports

```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

## 2. Load datasets

```python
file_covid = 'output_02_data_02_covid_cities_confirmed_cases.csv'
path_input_file_covid = os.path.join(os.getcwd(), '..', 
                                     'sprint_03_data_analysis', 'output', 
                                     file_covid)
df_covid = pd.read_csv(path_input_file_covid, delimiter=';')
del df_covid['Unnamed: 0']
df_covid['codmun'] = df_covid['codmun'].map(str).map(lambda codmun: codmun.split('.')[0])
df_covid.head()
```

```python
input_path_indicators_2 = os.path.join(os.getcwd(), '..', 'sprint_03_data_analysis', 'output')
input_file_indicators_2 = 'output_04_data_07_brazil_indicators_02_with_states_and_cities.csv'
df_indicators_2 = pd.read_csv(os.path.join(input_path_indicators_2, input_file_indicators_2)).drop('Unnamed: 0', axis=1)
df_indicators_2.head()
```

## 3. Join dataframes df_covid and df_indicators_2

```python
print(df_covid.shape)

li_index_covid_state_city = []

for state, city in tqdm(zip(df_covid['estado'], df_covid['municipio'])):
    index_state_city = state + '-' + city    
    li_index_covid_state_city.append(index_state_city)

arr_index_covid_state_city = np.array(li_index_covid_state_city)

df_covid.index = arr_index_covid_state_city

del li_index_covid_state_city
del arr_index_covid_state_city
```

```python
df_covid
```

```python
print(df_indicators_2.shape)

li_index_indicators_2_state_city = []

for state, city in tqdm(zip(df_indicators_2['Estado'], df_indicators_2['Municipio'])):
    index_state_city = state + '-' + city    
    li_index_indicators_2_state_city.append(index_state_city)

arr_index_indicators_2_state_city = np.array(li_index_indicators_2_state_city)

df_indicators_2.index = arr_index_indicators_2_state_city

del li_index_indicators_2_state_city
del arr_index_indicators_2_state_city
```

```python
df_indicators_2
```

```python
df_covid_indicators = df_covid.join(df_indicators_2)
```

```python
df_covid_indicators.head().iloc[1]
```

```python
del df_covid_indicators['Municipio']
del df_covid_indicators['Estado']
```

## 4. Save

```python
path_output = os.path.join(os.getcwd(), 'output')
file_output = 'output_05_covid19_2020-2021_and_indicators_2010.csv'
df_covid_indicators.to_csv(os.path.join(path_output, file_output))
```
